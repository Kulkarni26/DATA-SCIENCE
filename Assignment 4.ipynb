```python
import numpy as np
import pandas as pd
patient_df = pd.read_csv("Patient_Data.csv")
billing_df = pd.read_csv("Billing_Data.csv")


#1st Task
print("Patient Data Info: ")
print(patient_df.info())
print("\nBilling Data Info: ")
print(billing_df.info())

#2nd & 3rd Task
billing_cols= ['PatientID', 'Department', 'Doctor', 'BillAmount']
billing_data = patient_df[billing_cols].copy()
print("\nBilling Data:")
print(billing_data)

#3rd task
patient_cleaned = patient_df.drop(['ReceptionistID', 'CheckInTime'],axis=1)
print("\nAfter removing ReceptionistId and Checkin Time: ")
print(patient_cleaned)


#4th task
total_bill_dept = patient_df.groupby(['Department'])['BillAmount'].sum()
print("\nTotal Bill per Depatment: ")
print(total_bill_dept)

#5th task
patient_cleaned = patient_cleaned.drop_duplicates("PatientID")
print("\nAfter Removing Duplicates: ")
print(patient_cleaned)


#6th task
Mean_bill = patient_cleaned["BillAmount"].mean()                 #Finding mean of the total bill after removing duplicates

patient_fill = patient_cleaned.copy()                            #copying the data of patient_cleaned into patient_fill

patient_fill['BillAmount']=patient_fill['BillAmount'].fillna(Mean_bill)
t1 = np.round(patient_fill,1)                                    #round the decimal values to 1 didgit
print("\nAfter Replacing null values: ")
print(t1)



#7th task
merge_df = pd.merge(patient_fill,billing_df, on="PatientID")
print("\nAfter merging patient_cleaned and billing data: ")
print(merge_df)


#8th task
new_patients = pd.DataFrame({
    'PatientID': [106, 107],
    'Department': ['Cardiology', 'Orthopedics'],
    'Doctor': ['Dr. Shah', 'Dr. Mehta'],
    'BillAmount': [2500, 3200],
    'InsuranceCovered': [2000, 4000],
    'FinalAmount': [2800, 4200]
})
final_df = pd.concat([merge_df, new_patients], ignore_index=True)

print("\nFinal DataFrame after adding new patients: ")
print(final_df)


#9th task
patient_df_indexed = final_df.set_index('PatientID')
billing_df_indexed = billing_df.set_index('PatientID')

column_concatenated_df = pd.concat([patient_df_indexed, billing_df_indexed], axis=1)

print("\nDataFrame after column-wise concatenation: ")
print(column_concatenated_df)

column_concatenated_df.to_csv("Final_Hospital_Billing.csv", index=False)
print("\n✅ Final cleaned dataset saved as 'Final_Hospital_Billing.csv'")
```
