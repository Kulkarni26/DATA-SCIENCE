```python
# Exercise 1: Prime Numbers
num = int(input("Enter a number: "))
if num > 1:
    for i in range(2, num):
        if num % i == 0:
            print(num, "is not a prime number.")
            break
    else:
        print(num, "is a prime number.")
else:
    print("Number should be greater than 1.")
```

```python
# Exercise 2: Product of Random Numbers
import random

num1 = random.randint(1, 10)
num2 = random.randint(1, 10)

print(f"What is the product of {num1} and {num2}?")
user_answer = int(input("Your answer: "))

if user_answer == num1 * num2:
    print("✅ Correct!")
else:
    print(f"❌ Wrong! The correct answer is {num1 * num2}.")
```

```python
# Exercise 3: Squares of Even Numbers (choice: Even numbers)
print("Squares of even numbers between 100 and 200:")

for num in range(100, 201):
    if num % 2 == 0:
        print(f"{num}² = {num ** 2}")
```

```python
# Exercise 4: Hospital Billing System

room_charge_per_day = 2000
doctor_fee = 1500
lab_test_charge = 300

days_stayed = 4
num_tests = 3
medicine_cost = 2400

# Calculate total
total_before_discount = (room_charge_per_day * days_stayed) + doctor_fee + (lab_test_charge * num_tests) + medicine_cost

# Apply discount if applicable
if total_before_discount > 10000:
    discount = total_before_discount * 0.10
else:
    discount = 0

final_bill = total_before_discount - discount

print("----- Hospital Bill -----")
print(f"Room Charges ({days_stayed} days): ₹{room_charge_per_day * days_stayed}")
print(f"Doctor Fee: ₹{doctor_fee}")
print(f"Lab Tests ({num_tests} tests): ₹{lab_test_charge * num_tests}")
print(f"Medicine Charges: ₹{medicine_cost}")
print(f"Discount: ₹{discount}")
print(f"Total Bill: ₹{final_bill}")
```
