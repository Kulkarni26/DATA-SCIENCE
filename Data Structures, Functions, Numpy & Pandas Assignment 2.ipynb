```python
#1.	Create a list of 5 integers. Perform and print the result of the following operations: append, extend, insert, remove, pop, clear, index, count, sort, and reverse.
n1 = [10,20,30,40,50]
print("Original Numbers:",n1)

n1.append(60)       #Adding 60 into list
print("After appending (60) :",n1)

n1.extend([70, 80])  # Add multiple elements
print("After extend:", n1)

n1.insert(2, 25)  # Insert element at index 2
print("After insert:", n1)

n1.remove(40)  # Remove first occurrence of value 40
print("After remove:", n1)

popped = n1.pop()  # Remove and return the last element
print("After pop:", n1, " | Popped element:", popped)

index_30 = n1.index(30)  # Get index of value 30
print("Index of 30:", index_30)

count_20 = n1.count(20)  # Count occurrences of 20
print("Count of 20:", count_20)

n1.sort()  # Sort ascending
print("After sort:", n1)

n1.reverse()  # Reverse the list
print("After reverse:", n1)

n1.clear()  # Remove all elements
print("After clear:", n1)
```

```python
#7.	Given a number, check if it's positive, and if it is also even. If not positive, print if it's zero or negative.
number = 1
if number > 0:
    if number % 2 == 0:
        print("\nThe number is positive and even.")
    else:
        print("The number is positive but odd.")
elif number == 0:
    print("The number is zero.")
else:
    print("The number is negative.")
```

```python
import pandas as pd
import numpy as np

# 11. Creating a small DataFrame manually
data = {
    'Name': ['Amit', 'Priya', 'Rahul', 'Sneha', 'Vikas', 'Amit', 'Kiran', 'Meena', 'Raj', 'Tina'],
    'Age': [25, 30, 28, 22, 35, 25, 29, 24, 31, 27],
    'City': ['Pune', 'Mumbai', 'Delhi', 'Chennai', 'Pune', 'Pune', 'Delhi', 'Mumbai', 'Chennai', 'Pune'],
    'Salary': [50000, 60000, 55000, np.nan, 65000, 50000, 58000, 62000, 64000, np.nan]
}

df = pd.DataFrame(data)
print("Original DataFrame:\n", df)

# Display info and statistics
print("\n--- DataFrame Info ---")
print(df.info())

print("\n--- Statistical Summary ---")
print(df.describe())

# Select 'Name' and 'City' columns
print("\n--- Selected Columns (Name, City) ---")
print(df[['Name', 'City']])

# Drop 'City' column
df_dropped = df.drop('City', axis=1)
print("\n--- DataFrame after dropping 'City' ---")
print(df_dropped)

# Fill missing values in 'Salary' with mean
df['Salary'] = df['Salary'].fillna(df['Salary'].mean())
print("\n--- DataFrame after filling missing Salary values ---")
print(df)

# Remove duplicate rows
df_no_duplicates = df.drop_duplicates()
print("\n--- DataFrame after removing duplicates ---")
print(df_no_duplicates)
```
